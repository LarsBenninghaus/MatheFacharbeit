**Save Plots**

```python
numberGames = 50
martingal=Game()
alambert=Game()

for i in range(1,36):
    basic_path = './Diagramme/boxplot/'

    df = martingal.martingal(numberGames=numberGames, bet_start=1, ignoreBudget=True, real=True, path="./Data/Gewinnzahlen Roulette 26.02.-04.03..csv")

    plt.figure(figsize=(6,4))
    df.boxplot()

    plt.savefig(basic_path + 'martingal/versuch(' + str(i) + ').png', dpi=300, bbox_inches="tight")
    plt.close()

    df = alambert.alambert(numberGames=numberGames, betStart=1, ignoreBudget=True, real=True, path="./Data/Gewinnzahlen Roulette 26.02.-04.03..csv", unit=1)

    plt.figure(figsize=(6,4))
    df.boxplot()

    plt.savefig(basic_path + 'alambert/versuch(' + str(i) + ').png', dpi=300, bbox_inches="tight")
    plt.close()
```
